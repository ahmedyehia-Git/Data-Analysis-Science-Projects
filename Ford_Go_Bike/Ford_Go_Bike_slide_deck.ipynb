#### Filling null values:
* For member_birth_year we will use 'birth year mode' to fill the null value and for member_gender we will use 'other'
* For start_station_id, start_station_name, end_station_id and end_station_name we will drop these rows as it is 197 records out of 183412 total records so it will not affect our analysis

```python
df.member_gender.unique()
```

```python
birth_year_mode = df['member_birth_year'].mode()[0]
df['member_birth_year'].fillna(birth_year_mode, inplace = True)
```

>#### There are some illogical birth year records needed to be removed, the cut point should be 1932 i.e. users that have 90 years of old as of now

#### Examine and delete outliers in year of birth, those who are more than 90 years old 

```python
age_outliers = df.query('member_birth_year < 1932')
age_outliers.sample(5)
```

#### Remove age outliers

```python
df = df.query('member_birth_year >= 1932')
```

```python
df.info()
```

>#### We remove 167 records as birth year outliers; now the dataset contains 183048 records 

#### Change datatype of 'start_time' and 'end _time' to be datetime

```python
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
```

```python
df.info()
```

#### Add new columns for trip_start_hour & trip_weekday

```python
df['trip_start_hour'] = df['start_time'].dt.hour
df['trip_weekday'] = df['start_time'].dt.day_name()
```

```python
df.head()
```

#### Calculate the user age and add it as a new column

```python
import datetime
current_year = datetime.datetime.now().year
df['age'] = current_year - df['member_birth_year']
df.head()
```

```python
df.age.unique()
```

#### Add new column for trip

```python
df['trip'] = df['start_station_name'] + ' to ' + df['end_station_name']
df.head()
```

#### Add new column to calculate duration in min.

```python
df['duration_min'] = df['duration_sec']/60
```

```python
df.duration_min.min() , df.duration_min.max()
```

#### Change member_gender, & bike_share_for_all_trip to categorical datatype

```python
member_gender = ['Male', 'Female', 'Other']
gender = pd.api.types.CategoricalDtype(ordered = True, categories = member_gender)
df['member_gender'] = df['member_gender'].astype(gender)
```

```python
user_loyalty = ['No', 'Yes']
loyalty = pd.api.types.CategoricalDtype(ordered = True, categories = user_loyalty)
df['bike_share_for_all_trip'] = df['bike_share_for_all_trip'].astype(loyalty)
```
